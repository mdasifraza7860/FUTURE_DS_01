## Library Import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
df=pd.read_excel(r"C:\Users\mrasi\Downloads\Sales_data.xlsx")
```

```python
df.head()
```

## KPI Analysis

```python
total_revenue = df["Sales"].sum()
print("Total Revenue:", total_revenue)
```

## Revenue Trend(Monthly)

```python
df["Month"] = df["Date"].dt.to_period("M")

monthly_sales = df.groupby("Month")["Sales"].sum()
monthly_sales.plot(kind="line", marker="o", title="Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

## Top Selling Products

```python
product_sales = df.groupby("Product")["Sales"].sum().sort_values(ascending=False)

product_sales.plot(kind="bar", title="Top-Selling Products")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

## Category Performance

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales.plot(kind="bar", title="Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

## Regional Performance

```python
region_sales = df.groupby("Region")["Sales"].sum()

region_sales.plot(kind="bar", title="Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

## Key Insights

* Electronics category generates the highest revenue among all categories.

* Laptop is the top-selling product, contributing the largest share of total sales.

* East region shows the strongest sales performance.

* Sales increased from January to February, indicating positive revenue growth.

## Business Recommendations

* Focus marketing efforts on Electronics products to maximize revenue.

* Increase inventory for high-demand products like Laptops.

* Expand sales strategies in high-performing regions such as the East.
